```python
import pandas as pd
from os import path

train_dir = './data/train'
val_dir = './data/val'
test_dir = './data/test'

TRAIN_BEHAVIORS: str = path.join(train_dir, 'train_behaviors.tsv')
TRAIN_NEWS: str = path.join(train_dir, 'train_news.tsv')
RANDOM_SEED: int = 310505030
```

```python
behaviors = pd.read_csv(
  TRAIN_BEHAVIORS, 
  delimiter='\t',
  index_col=0,
  # names=['impression_id', 'user', 'time', 'clicked_news', 'impressions'],
  )
behaviors = behaviors.sample(frac=1, random_state=RANDOM_SEED)
train_behaviors = behaviors[:int(len(behaviors) * 0.9)]
valid_behaviors = behaviors[int(len(behaviors) * 0.9):]

valid_behaviors
```
